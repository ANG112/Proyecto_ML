## TRANSFORMACIÓN DE CATEGÓRICAS

### One Hot Encoding

#### Variables a aplicar OHE

```python
var_ohe = ['marital_status']
```

#### Instanciar

```python
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
```

#### Entrenar y aplicar

```python
cat_ohe = ohe.fit_transform(cat[var_ohe])
```

#### Guardar como dataframe

```python
cat_ohe = pd.DataFrame(cat_ohe, columns = ohe.get_feature_names_out(), index=cat.index)
```

### Ordinal Encoding

#### Variables a aplicar OE

```python
var_oe = ['education']
```

#### Orden de los valores de las variables

```python
#Orden de la primera variable
orden_formacion = ['Basic', '2n Cycle', 'Graduation', 'Master', 'PhD']
```

#### Instanciar

```python
oe = OrdinalEncoder(categories = [orden_formacion],
                    handle_unknown = 'use_encoded_value',
                    unknown_value = 10)
```

#### Entrenar y aplicar

```python
cat_oe = oe.fit_transform(cat[var_oe])
```

#### Guardar como dataframe

```python
#Añadir sufijos a los nombres
nombres_oe = [variable + '_oe' for variable in var_oe]

#Guardar como dataframe
cat_oe = pd.DataFrame(cat_oe, columns = nombres_oe,index=cat.index)
```

## UNIFICAR DATASETS TRANSFORMADOS

### Lista de los dataframes generados

```python
dataframes = []
dataframes.extend(value for name, value in locals().items() if name.startswith('cat_') or name.startswith('num_'))
dataframes.append(num)
```

### Unir todos los dataframes

```python
df = pd.concat(dataframes, axis = 1)
```

#### Variables a reescalar con Min-Max

```python
var_mms = df[['education_oe',
       'income', 'recency', 'mntwines', 'mntfruits',
       'mntmeatproducts', 'kidhome', 'teenhome','mntfishproducts', 'mntsweetproducts',
       'mntgoldprods', 'numdealspurchases', 'numwebpurchases',
       'numcatalogpurchases', 'numstorepurchases', 'numwebvisitsmonth',
       'age', 'household_members', 'total_amount', 'total_purchase', 'median_amount_purchase', 'total_cmp',
       'total_%_cmp']].columns
```

```python
var_mms
```

#### Instanciar

```python
mms = MinMaxScaler()
```

#### Entrenar y aplicar

## UNIFICAR DATASETS REESCALADOS

### Crear una lista con los dataframes a incluir en el tablón analítico
